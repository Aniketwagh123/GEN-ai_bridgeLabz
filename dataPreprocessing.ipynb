```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data/data_preprocessing.csv")
df
```

```python
df['Country_num'] = df['Country'].map({'France': 0, 'Spain': 1, 'Germany': 2})
df['Purchased_num'] = df['Purchased'].map({'No': 0, 'Yes': 1})
df
```

```python
df['Age_mean_filled'] = df['Age'].fillna(df['Age'].mean())
df['Salary_mean_filled'] = df['Salary'].fillna(df['Salary'].mean())

df['Age_median_filled'] = df['Age'].fillna(df['Age'].median())
df['Salary_median_filled'] = df['Salary'].fillna(df['Salary'].median())
df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))
```

```python
# Prepare data for line plots
age_data = df[['Age', 'Age_mean_filled', 'Age_median_filled']].dropna()
salary_data = df[['Salary', 'Salary_mean_filled', 'Salary_median_filled']].dropna()
```

```python
# Covariance matrix
cov_matrix = df[["Age", "Salary", 'Age_mean_filled', 'Age_median_filled', 'Salary_mean_filled',"Salary_median_filled"]].cov()

cov_matrix
```
